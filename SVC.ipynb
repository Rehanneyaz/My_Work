```python
# lets split the data into dependent and independent features

x= data.drop("quality",axis=1)
y = data["quality"]
```

```python
x
```

```python
y
```

```python
# lets split into train and test data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# lets bring all values on same scale by using standard scaler

from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
x_train= scaler.fit_transform(x_train)
```

```python
x_train
```

```python
# lets standardised the test data by using same mean and standard deviation of train data
# to avoid data leakage
x_test= scaler.transform(x_test)
```

```python
x_test
```

```python
# model building

from sklearn.svm import SVC
model = SVC()
```

```python
# lets fit the data

model.fit(x_train, y_train)
```

```python
# lets check model training accuracy
model.score(x_train,y_train)
```

```python
# lets check model test accuracy

y_pred= model.predict(x_test)
```

```python
from sklearn. metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# lets check by using logistic regression in which we apply one verses rest
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model2=  lr.fit(x_train,y_train)
```

```python
# lets check the accuracy
lr_pred= model2.predict(x_test)
```

```python
lr_pred
```

```python
from sklearn. metrics import accuracy_score
accuracy_score(y_test,lr_pred)
```
